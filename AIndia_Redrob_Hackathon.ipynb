```python
import json

candidate_path = "/content/data/[PUB] India_runs_data_and_ai_challenge/India_runs_data_and_ai_challenge/candidates.jsonl"

candidates = []

with open(candidate_path,"r") as f:

    for line in f:

        candidates.append(json.loads(line))

print("Total candidates:",len(candidates))
```

```python
import pandas as pd

rows = []

for c in candidates:

    p = c["profile"]

    r = c["redrob_signals"]

    row = {

        "candidate_id": c["candidate_id"],

        "headline": p["headline"],

        "summary": p["summary"],

        "location": p["location"],

        "country": p["country"],

        "current_title": p["current_title"],

        "current_company": p["current_company"],

        "current_industry": p["current_industry"],

        "years_of_experience": p["years_of_experience"],

        "current_company_size": p["current_company_size"],


        # behavioral

        "open_to_work":

            r["open_to_work_flag"],

        "github_score":

            r["github_activity_score"],

        "recruiter_response_rate":

            r["recruiter_response_rate"],

        "interview_completion_rate":

            r["interview_completion_rate"],

        "offer_acceptance_rate":

            r["offer_acceptance_rate"],

        "search_appearance":

            r["search_appearance_30d"],

        "saved_by_recruiters":

            r["saved_by_recruiters_30d"],

        "profile_views":

            r["profile_views_received_30d"],

        "notice_period":

            r["notice_period_days"],

        "willing_to_relocate":

            r["willing_to_relocate"]

    }

    rows.append(row)

df = pd.DataFrame(rows)

df.head()
```

```python
def get_title_score(title):

    title = str(title).lower()


    if "senior ai engineer" in title:
        return 100

    elif "lead ai engineer" in title:
        return 95

    elif "staff machine learning engineer" in title:
        return 95

    elif "senior machine learning engineer" in title:
        return 92

    elif "senior applied scientist" in title:
        return 90

    elif "ai research engineer" in title:
        return 88

    elif "machine learning engineer" in title:
        return 85

    elif "applied ml engineer" in title:
        return 85

    elif "ml engineer" in title:
        return 82

    elif "ai engineer" in title:
        return 80

    elif "senior nlp engineer" in title:
        return 78

    elif "nlp engineer" in title:
        return 75

    elif "senior data scientist" in title:
        return 72

    elif "data scientist" in title:
        return 65

    elif "senior software engineer" in title and "ml" in title:
        return 60

    elif "data engineer" in title:
        return 40

    elif "senior data engineer" in title:
        return 45

    elif "data analyst" in title:
        return 20

    else:

        return 0
```

```python
SKILL_WEIGHTS = {

    # Core AI

    "LLM":15,

    "Large Language Models":15,

    "Generative AI":15,

    "Embeddings":14,

    "Retrieval":14,

    "Ranking":14,

    "RAG":13,

    "NLP":12,

    "Prompt Engineering":12,

    "LangChain":12,

    # Frameworks

    "Python":8,

    "PyTorch":8,

    "Transformers":8,

    "Machine Learning":8,

    "Deep Learning":8,

    # Vector DBs

    "Milvus":10,

    "Pinecone":10,

    "Weaviate":10,

    "Vector Search":10,

}
```

```python
def get_skill_score(skills):

    score = 0

    candidate_skills = {

        s["name"].lower()

        for s in skills

    }


    for skill,weight in SKILL_WEIGHTS.items():

        if skill.lower() in candidate_skills:

            score += weight


    return score
```

```python
def get_exp_score(exp):

    if 5 <= exp <= 9:
        return 100

    elif exp == 4 or exp == 10:
        return 80

    elif exp == 3 or exp == 11:
        return 60

    else:
        return 20
```
